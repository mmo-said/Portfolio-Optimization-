# 2.1 Background

Where they compare the 60/40 portfolio with a riskier portfolio say 70/30 but with tail hedges. He shows that by taking more equity risk, the capability of harvesting more equity risk premium that can pay for tail heding. He also shows that by the comparison of unhedged 60/40 portfolio with the 70/30 hedged portfolio. that the hedged 70/30 portfolio has a greater volatility however this volatility comes in two forms, thanks to the put protection, it provides a similar downside volatility as the 60/40 however with a much greater upside volatility because of the 70/30 split, hence it yields a higher expected rate of return with the same risk premium rate. They also illustrated that the oppourtunity to increase risk exposures in a tail event improves the return distribution of most investment portfolios. He showed that the shadow value of a tail hedging program is positive [3]. 

The paper conculed [3] that tail hedging includes a risk premium, hence the expected return on for example equity options is less than zero, having a negative expected return implied that as a standalone allocation in a portfolio, the tail hedge is a poor choice as it has a negative expected return. However on relative basis and paired with another strategy, the tail hedge allows the investor to scale up thier exposure to risky assets that have a gain in risk premium, thus the investor should compare the relative risk premium gained from investing in risky assets to that lost form investing in tail hedges, hence mainting an overall risk target is crucial.

An added benefit of tail hedging is a noticeable improvement in risk adjusted return measured in terms of the Sharpe ratio provided by the cash based strategies; the relatively poor performance of the options based strategy is found to occur during the 2004–2015 period and is mostly due to a significant increase in the price of puts with respect to calls which acts as a performance drag.

# Strangles

There are many types of strangles, a basic long strangle consists in the simultaneous purchase of a put with strike K1 and a call with strike K2 with the same maturity. The strategiy has to satisfy K1 > K2.

The investor is betting that there will be a large price move, but is uncertain whether it will be an increase or a decrease. The profit pattern obtained with a strangle depends on how close together the strike prices are. The farther they are apart, the less the downside risk and the farther the stock price has to move for a profit to be realized.

As shown by the diagram below is the stock price stays within the highlighted red then your payoff is zero. However if the stock price moves beyond your k1 or k2 strike then your option will have a payoff. 

Total payoff will be: 

if ST <= K1 then K1 - ST 

if ST >= K2 then ST- K2 

if K1 < ST < K2 then payoff is zero.

# Spreads

A spread is a options trading strategy which involves taking a position in two or more options of the same type (two or more call or two or more puts).

An example can a call spread or a bull spread. This strategy consists in the purchase of a long call with strike K1 and short a call with a higher strike strike K2 (K1 < K2) with the same maturity. 

Example: Suppose that K1 is the strike price of the call option bought, K2 is the strike price of the call option sold, and ST is the stock price on the expiration date of the options. 
If the stock price does well and is greater than the higher strike price, the payoff is the difference between the two strike prices, or K2 - K1.

If the stock price on the expiration date lies between the two strike prices, the payoff is ST - K1.

If the stock price on the expiration date is below the lower strike price, the payoff is zero. [12]

A bull spread strategy limits the investor's upside as well as downside risk.

These instruments are popular in the catastrophe options market: the fact that the potential loss is bounded both for the buyer and the seller is a feature meant to generate liquidity. This type of option strategy is a chapped strategy, meaning the payoff is bounded by the spread between K2-K1.

```python
option_K1 = {'op_type': 'c', 'strike': 800, 'tr_type': 'b', 'op_pr': 200}
option_K2 = {'op_type': 'c', 'strike': 1100, 'tr_type': 's', 'op_pr': 100}

op_list=[option_K1, option_K2]

op.multi_plotter(spot=900,spot_range=100, op_list=op_list)
```

# Covered Calls 

Is a portfolio that is equivalent to holding a long position in an asset and shorting a call option of the same underlying. The investor's long position in the asset is the cover in case the the underlying stock goes above the strike of the short call, in that case the loss incurred from the short call will be covered from the long position in the underlying. Essentially the investor's long position is the cover in the call option buyer chooses to exercise. 

The payoff is as follows:

CoveredCall(Payoff) (K, T) = ST - S0 − max [0, ST − K] + premium 

where ST is the underlying price at expiration.
S0 is the underlying price at time 0 
K is the strike price. 



This is a popular strategy amongst fund managers it enables them to get protection from a small decrease in the stock but also increase their portfolio returns in form income from the short call option premium.

An example is, say a trader holding a position in a specific company and he believes that this company will increase in price over the long term, hence he or she is not willing to sell however the trader thinks that the stock will decline in the short term, hence the trader decides to sell a short maturity call on the underlying and use the premium gained from short selling to compensate for the potential losses in the long position.

In the case that the expectation is are correct, the stock price goes slightly down in the short, hence the short call will be in the money and the premium gained can then be balanced by the loss in the long position.

Alternatively, if the stock goes up significantly then his long position in the underlying will be positive and it will cover for the loses in the short call.

This is considered a short term hedge ie a neutral strategy. This strategy is not useful if you have a strong bullish or bearish outlook on a stock. 

```python
s0=189 # Initial stock price 
k=195;c=6.30; # Strike price and Premium of the option
shares = 100 
sT = np.arange(0,2*s0,5) # Stock Price at expiration of the Call

y1= (sT-s0) * shares #  PNL

# Payoff from a Short Call Option
y2 = np.where(sT > k,((k - sT) + c) * shares, c * shares)

# Payoff from a Covered Call
y3 = np.where(sT > k,((k - s0) + c) * shares,((sT- s0) + c) * shares )

# Create a plot using matplotlib 
fig, ax = plt.subplots()

ax.spines['top'].set_visible(False) # Top border removed 
ax.spines['right'].set_visible(False) # Right border removed
ax.spines['bottom'].set_position('zero') # Sets the X-axis in the center
ax.tick_params(top=False, right=False) # Removes the tick-marks on the RHS

plt.plot(sT,y1,lw=1.5,label='Long Stock')
plt.plot(sT,y2,lw=1.5,label='Short Call')
plt.plot(sT,y3,lw=1.5,label='Covered Call')

plt.title('Covered Call') 
plt.xlabel('Stock Prices')
plt.ylabel('Profit/loss')

plt.grid(True)
plt.axis('tight')
plt.legend(loc=0)
plt.show()

#[11]
```

# 3.2 Measuring Expected Return and Risk for Mean Variance Optimization. 

Harry Markowitz Mean Variance Optimization theory is a mathematical programming and statistical analysis that aims to arrange an optimal allocation of assets within the portfolio based on expected return, variance and correlation between assets within a specific portfolio [10][7].

The theory states that an efficient frontier of optimal portfolio can be constructed to offer the investor the maximum possible expected return for a given level of risk or variance. The theory also assumes that investors estiamte risk on the basis of the variability of expected returns hence an investor is looking to achieve higher expected return with the lowest possible variability of expected returns or variance of returns ie risk. 

Now lets begin by calculating the expected return. 

Our expected return will be computed using the Annualised Percentage Rate (APR) [7]. 

Formula is as follows: 

Expected Return = $$ E[r] = E[P(n) / P0 - 1] $$

where $$ r = P(n)/ P0 - 1 $$ 

where n is time = 1, 2, 3, 4 .... n 

The advantage for using that is that it checks for profitability in every year. 


```python
APR = Portfolio_returns.groupby([Portfolio_returns.index.year]).agg('sum')
APR_avg = APR.mean()

APR
```

# 3.3 Portfolio CoVariance Matrix & Correlation



Covariance matrix can be used to determine which asset to include in the portfolio. The matrix is a statistical measure of the directional relationship between two asset price. Covariance matrix is one of the components used in modern portfolio theory to determine the overall risk of a portfolio[8]. It measures the extent in which two assets move together, the direction of the relationship between the two variables[10]. 

A positive covariance means asset prices move in the same direction while a negative covariance means assets move in the opposite direction.

# Efficient Frontier

The Efficient Frontier, shows all the combinations of allocation for a given portfolio. In our case we will use the data from Portfolio_returns and simulate 10000 portfolio and plot them to observe all the possible combinations

```python
p_ret = [] # empty list for portfolio returns 
p_vol = [] # empty list for portfolio volatility 
p_weights = []  #empty array for asset weights 

num_assets = len(Portfolio_returns.columns)
num_portfolios = 10000

#[13]
```

```python
for portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)
    returns = np.dot(weights, APR_avg)
    
    p_ret.append(returns)
    var = Portfolio_cov_matrix .mul(weights, axis=0).mul(weights, axis=1).sum().sum()# Portfolio Variance
    sd = np.sqrt(var) # Daily standard deviation
    ann_sd = sd*np.sqrt(250) # Annual standard deviation = volatility
    p_vol.append(ann_sd)
```

```python
data = {'Returns':p_ret, 'Volatility':p_vol}

for counter, symbol in enumerate(Portfolio.columns.tolist()):
    #print(counter, symbol)
    data[symbol+' weight'] = [w[counter] for w in p_weights]
```
